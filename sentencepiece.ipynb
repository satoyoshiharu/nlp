```
import codecs
import re
import requests
from bs4 import BeautifulSoup
import codecs
import urllib.parse as parser
import urllib.request as request

url = "https://ja.wikipedia.org/wiki/"
keyword_list = [
"ロボット","ロケット","コンピュータ","人工知能","自動車","宇宙","機械","道具","計算","装置",
"プログラム","メモリ","CPU","GPU","クラウド","サーバー","クライアント",
"情報","構造","ネットワーク","性能","演算","記憶",
"太陽","月","彗星","惑星","銀河系","水星","金星","火星","木星","土星","冥王星","天王星",
"人間","目","口","耳","舌","肌","指","手","足","頭","首","声","腹","背","胃","腸","髪","血液","免疫",
"動物","植物","猫","犬","鳩","シジュウカラ","スズメ","ネズミ","魚","鳥","昆虫","猿","烏","ミミズ","ナメクジ","蟻","熊","クジラ","イルカ",
"蛇","ウナギ","トンボ","カマキリ","ミツバチ","ハエ","カブトムシ","豚","牛","鶏",
"日本","東京","横浜","北海道","東北","関東","関西","北陸","中部","北陸","近畿","四国","九州","沖縄",
"アジア","アメリカ","中国","韓国","台湾","香港","北朝鮮","アフリカ","中近東","オセアニア",
"電気","信号","電波","重力","原子","分子","素粒子","量子",
"紙","プラスチック","鉄","金","銀","銅","アルミニウム","ビニール","土","火","水","ガラス",
"聖徳太子","坂本龍馬","徳川家康","源義経","夏目漱石","宮沢賢治","斎藤茂吉",
"サッカー","野球","ラグビー","プロレス","テニス","バレーボール","ホッケー","スキー","スケート",
]
corpus = []
for keyword in keyword_list:
    response = requests.get(url + keyword)
    html = response.text
    soup = BeautifulSoup(html, 'lxml')
    for p_tag in soup.find_all('p'):
        text = "".join(p_tag.text.strip().split(" "))
        p_tags = soup.find_all('p')
        corpus.append(text)

print(len(corpus))
print(corpus[0:5])
with codecs.open("pwiki.txt", "w", "utf-8") as f:
    f.write("\n".join(corpus))
print(*corpus, sep="\n", file=codecs.open("wiki.txt", "w", "utf-8"))
```

```
import sentencepiece as spm
spm.SentencePieceTrainer.Train('--input=wiki.txt, --model_prefix=sentencepiece --character_coverage=0.9995 --vocab_size=50000 --model_type=unigram')
```

```
sp = spm.SentencePieceProcessor()
sp.Load("sentencepiece.model")
print(sp.EncodeAsPieces('横浜工場で、情報処理装置を開発する'))
print(sp.EncodeAsPieces('ペットの犬がしっぽを振りながら迎えに駆け寄ってきた。'))
```
