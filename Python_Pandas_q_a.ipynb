##問題1 Pandas参照
以下のデータフレームから、役割のリストを作れ。
```
import pandas as pd
df = pd.DataFrame({
    '名前' : ['田中', '山田', '高橋'],
    '役割' : ['営業部長', '広報部', '技術責任者'],
    '身長' : [178, 173, 169]
    })
```
出力は、
```
0     営業部長
1      広報部
2    技術責任者
Name: 役割, dtype: object
```

````
import pandas as pd
df = pd.DataFrame({
    '名前' : ['田中', '山田', '高橋'],
    '役割' : ['営業部長', '広報部', '技術責任者'],
    '身長' : [178, 173, 169]
    })
print(df['役割'])
print(df.iloc[0])
````

##問題2 Pandas行参照
上記のデータフレームから、第0番目の行を取り出せ。

出力は、以下とする。
```
名前      田中
役割    営業部長
身長     178
Name: 0, dtype: object
```



````
print(df.loc[0])
````

##問題3 Pandas条件検索
問題1のデータフレームから、役割が技術責任者であるデータを取り出せ。
出力は以下とする。

```
   名前     役割   身長
2  高橋  技術責任者  169
```



````
print(df[df['役割']=='技術責任者'])
print(df.query('役割=="技術責任者"'))
````

##問題4 Pandasソート
問題6のデータを、身長の昇順にソートせよ。

````
print(df.sort_values('身長', ascending=True))
````
