##課題05. n-gram
与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ．この関数を用い，”I am an NLPer”という文から単語bi-gram，文字bi-gramを得よ．

###ヒント
n-gramというのは、要素の連続する並びデータ。

gramは、文法的要素の意味。
n-gramのnは、1(uni)、2(bi)、3(tri)だったりする。
また、要素は文字だったり、単語だったりする。

例:
  abcdefgの文字unigramは、a,b,c,d,e,f,g

  abcdefgの文字bigramは、(a,b),(b,c),(c,d),(d,e),(e,f),(f,g)

  abcdｅｆｇの文字trigramは、(a,b,c),(b,c,d),(c,d,e),(d,e,f),(e,f,g)

  "abc de fg hij"の単語unigramは、abd, de, fg, hij

  "abc de fg hij"の単語bigramは、(abd, de), (de, fg), (fg,hij)

  "abc de fg hij"の単語trigramは、(abc, de, fg), (de,fg,hij)
