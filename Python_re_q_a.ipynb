##問題2
上記のログから、python3とCRONとsystemdの出力以外の行を取り出せ。

````
import re
regexp = r'python3|CRON|systemd'
lines = str.split('\n')
for l in lines:
  if len(re.findall(regexp, l)) == 0:
    print(l)
````

##問題3
以下のurlから、プロトコル部分を抜き出せ。出力は、以下となる。

```
['https']
['wss']
['apps']
```



````
import re
regexp = r'^([^:]+)'
print(re.findall(regexp, 'https://abc.com/def/'))
print(re.findall(regexp, 'wss://xyz.com:8080/'))
print(re.findall(regexp, 'apps://myapplication/'))
````

##問題4
urlから、ポート番号を除くドメイン名を抜き出せ。出力は、以下となる。

```
['abc.com']
['xyz.com']
['myapplication']
```

````
regexp = r'\/\/([^\/:]+)'
print(re.findall(regexp, 'https://abc.com/def/'))
print(re.findall(regexp, 'wss://xyz.com:8080/'))
print(re.findall(regexp, 'apps://myapplication/'))
````
