##課題02. 「パトカー」＋「タクシー」＝「パタトクカシーー」
「パトカー」＋「タクシー」の文字を先頭から交互に連結して文字列「パタトクカシーー」を得よ

###ヒント
リストの内包表記で、リストを表現で指定できる。 

[式 for 任意の変数名 in 繰り返しデータ] 

[式 for 任意の変数名 in 繰り返しデータ if 条件]

[式 if 条件 else 式 for 任意の変数名 in 繰り返しデータ]

https://note.nkmk.me/python-list-comprehension/

リストの内包表記は、for文などのループより高速なので、内包表記で書ける場合はできるだけ内包表記を使いましょう。

リスト要素は以下で文字列連結できる：
セパレータ文字列.join（繰り返しデータ）

複数のリストからまとめて、1個づつ取り出すのにZIP関数が利用されます。
https://note.nkmk.me/python-zip-usage-for/

###解答例

```
s1 = 'パトカー'
s2 = 'タクシー'
a = ''
for i in range(len(s1)):
  a += s1[i]
  a += s2[i] 
print('forを使った例:\t', a)

a_list = [s1[i]+s2[i] for i in range(len(s1))]
a = ''.join(a_list)
print('\nリストの内包表記を使った例:\t', a)

print("\nzipを使った例:")
z = zip(s1,s2) # zipは、複数の繰り返しオブジェクトから1個づつ取り出してタプルで返すイテレータ
# zipはイテレータなので、ままprintできない
print("zip イテレータのダンプ:")
for i, e in enumerate(zip(s1,s2)): #イテレータは先頭に戻せないので、いちいち作る
  print("\t", i, e)

a_list = [i+j for i, j in zip(s1,s2)] #イテレータからs1とs2の対応する文字のタプルを取り出し、要素を+で連結して、リストにしている
print('イテレータの中の要素を連結してリストへ:\t', a_list)

a = ''.join(a_list) # リストの要素を''でつなぐ
print('リストの要素を空文字でつなぐ:\t', a)
```

##課題04. 元素記号
“Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.”という文を単語に分解し，1, 5, 6, 7, 8, 9, 15, 16, 19番目の単語は先頭の1文字，それ以外の単語は先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）への連想配列（辞書型もしくはマップ型）を作成せよ．

###ヒント
enumerate関数で、繰り返しデータのIndexと値を一緒に取り出すことができます。https://note.nkmk.me/python-enumerate-start/　

辞書は、キーと値のペアの格納する連想配列です。テキスト処理では多用します。https://note.nkmk.me/python-dict-create/　

定義は
辞書 = {キー：値,…}

値の参照は
辞書[‘キー’]

挿入は
辞書[‘キー’]=値

辞書型に関するループは、リストと異なる書き方するので注意　https://www.python.ambitious-engineer.com/archives/186　

###解答例

```
s = "Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can."
d = {}
words = s.split()
filter = [1,5,6,7,8,9,15,16,19]

for i in range(len(words)):
  if i+1 in filter:
    d[words[i][:1]] = i+1
  else:
    d[words[i][:2]] = i+1
print(d)

for i, word in enumerate(words):
  if i+1 in filter:
    d[word[:1]] = i+1
  else:
    d[word[:2]] = i+1
print(d)

# [式 if 条件 else 式 for 任意の変数名 in 繰り返しデータ] の例
list = [(word[:1], i+1) if i+1 in filter 
  else (word[:2], i+1) 
  for i, word in enumerate(words)]
d = dict(list)
print(d)
```

##課題05. n-gram
与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ．この関数を用い，”I am an NLPer”という文から単語bi-gram，文字bi-gramを得よ．

###ヒント
n-gramというのは、要素の連続する並びデータ。

単語bigramなら、(n-gram, と),(と,いう),(いう,のは),…

文字Bigramなら、(n,-),(-,g),(g,r),…

###解答例

```
print('関数使わない版')
chars = "I am an NLPer" # 文字列は実は文字のリストなので、以下、単語リストと全く同じロジックを使える
words = chars.split() # 空白を境界として文字列分割
print('\t単語に分けてリスト: \t', words)
print('\t単語バイグラム:\t', [(words[i], words[i+1]) for i in range(len(words)-1)]) # 相並ぶ二つの単語を拾ってリストに
print('\t文字バイグラム:\t', [(chars[i], chars[i+1]) for i in range(len(chars)-1)]) #元のsから、相並ぶ2文字を拾ってリストに

print('関数使う版')
def ngram(data, n): 
# リストdataからn個の相並ぶ要素を拾って、tupleにして、それをリストにして返す
  # for版
  print(f'\t\tngram関数への引数, data={data}, n={n}')
  l = []
  for i in range(len(data)-n+1):
    l.append(tuple(data[i:i+n])) # 位置iから一i+nまでの部分を取り出しtupleへ変換、それをリストにアペンド
  return l
print('\tfor版')
print('\t\t', ngram(words, 2)) 
print('\t\t', ngram(chars, 2))

def ngram(data, n): 
# リストdataからn個の相並ぶ要素を拾って、tupleにして、それをリストにして返す
  # 内包表記版
  print(f'\t\tngram関数への引数, data={data}, n={n}')
  return [tuple(data[i:i+n]) for i in range(len(data)-n+1)] # 位置iから位置i+nまでの部分を取り出しtupleへ変換、それをリストにする
print('\t内包表記版')
print('\t\t', ngram(words, 2))
print('\t\t', ngram(chars, 2))
```

```
# zipを使うと
chars = "I am an NLPer"

def ngram(data, n):
  print(f'ngram関数への引数, data={data}, n={n}')
  
  # n=2ならrange(n)は0と1。dataから位置0から始まる部分と位置1から始まる部分(ずらした部分)を取り出す。
  # 以下で、試しにn=3としてみて違いを見てみてください。
  pairs = [data[i:] for i in range(n)] 
  print('zurasi pairs: ', pairs)
  
  # *はリストの要素を取り出すアンパックオペレータ。pairsはリストのリストなので、*pairsは内側のリストを取り出す。
  print('unpack: ' , *pairs)
  # 各リストの対応する要素を組みにしてtupleにし、そのリストを返す
  return list(zip(*pairs))

words = chars.split()
print('単語bigram', ngram(words, 2))
print('文字bigram', ngram(chars, 2))
```

##課題08. 暗号文 [難易度高]
与えられた文字列の各文字を，以下の仕様で変換する関数cipherを実装せよ.

*  英小文字ならば(219 - 文字コード)の文字に置換
*   その他の文字はそのまま出力

この関数を用い，英語のメッセージを暗号化・復号化せよ．


###ヒント
文字コード(ユニコード文字位置を表す16進数のint)を得るのはord(文字)。
文字コードから文字を得るのは chr(コード位置)。
https://note.nkmk.me/python-chr-ord-unicode-code-point/

f(n)=C-nとする。f(f(n))=C-(C-n)=n なので、C-n という関数を2回かければ元の数値になる。

###解答例

```
def cipher(str):

  # forの場合
  out = []
  for l in str:
    if l.islower():
      out.append(chr(219-ord(l)))
    else:
      out.append(l)
  print('cipher: for版', out)

  # 内包表記　[式 if 条件 else 式 for 任意の変数名 in 繰り返しデータ]
  out = [chr(219 - ord(l)) if l.islower() else l for l in str]
  print('cipher: 内包表記版',out)

  return ''.join(out)

print('abCの暗号化: ', cipher('abC'))
print('abCを暗号化したあと複合化: ',cipher(cipher('abC')))
```

##課題09. Typoglycemia
スペースで区切られた単語列に対して，各単語の先頭と末尾の文字は残し，それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ．ただし，長さが４以下の単語は並び替えないこととする．適当な英語の文（例えば”I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .”）を与え，その実行結果を確認せよ

###ヒント
リストから、ランダムに選びます https://note.nkmk.me/python-random-choice-sample-choices/　

ランダムに選んだリスト=random.sample(元リスト,取得要素数)

###解答例

```
s = "I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind ."
words = s.split()

import random
def shuffle(str):
  r = random.sample(str, len(str))
  return(''.join(r))

# forの場合
o = []
for w in words:
  if len(w)<=4: 
    o.append(w)
  else:
    o.append(w[:1] + shuffle(w[1:-1]) + w[-1:])
print('for版', o)

# 内包表記　[式 if 条件 else 式 for 任意の変数名 in 繰り返しデータ]
o = [w if len(w)<=4 else w[:1] + shuffle(w[1:-1]) + w[-1:] for w in words]
print('内包表記版', o)

print(' '.join(o))
```
